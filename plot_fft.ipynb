```python
import numpy as np, matplotlib.pyplot as plt
```

```python
#define x(t) as a pulse function. 
num_t = 100
xt = np.zeros(num_t) 
xt[:10] = 1
plt.plot(xt)
plt.xlabel("Plot of x(t)")
plt.xlabel("Sample index")
plt.ylabel("Value of x(t)")
```

```python
fxt = np.fft.fft(xt)
# First plot is the real part
plt.subplot(1, 2, 1)
plt.plot(np.real(fxt))
plt.title("Plot of real part of FT of x(t)")
plt.ylabel("Magnitude")
plt.xlabel("Frequency Index")
# FSecond plot is the imaginary part
plt.subplot(1, 2, 1)
plt.plot(np.imag(fxt))
plt.title("Plot of imaginary part of FT of x(t)")
plt.ylabel("Magnitude")
plt.xlabel("Frequency Index")
```
